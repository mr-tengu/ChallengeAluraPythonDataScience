# 1. Análisis de facturación



Calculamos la suma de los precios de todos los productos vendidos en cada tienda

```python
import matplotlib.pyplot as plt
facturacion = {}
for nombre, df in tiendas.items():
    facturacion[nombre] = df['Precio'].sum()

facturacion_df = pd.DataFrame(list(facturacion.items()), columns=['Tienda', 'Facturación Total']).sort_values('Facturación Total', ascending=False)
display(facturacion_df)

facturacion_df.plot(kind='bar', x='Tienda', y='Facturación Total', title='Facturación Total por Tienda', legend=False, color='skyblue')
plt.ylabel('Total (Moneda local)')
plt.show()
```
